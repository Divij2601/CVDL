```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('C:/Users/divij/Academics/CVDL/image.jpg')
```

```python
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap= 'gray')
grayf = gray.copy()
sft = gray.copy()
```

```python
row,col = gray.shape
for r in range(row):
    for c in range(int(col/2)):
        temp = grayf[r][c]
        grayf[r][c] = grayf[r][col-c-1]
        grayf[r][col-c-1] = temp
```

```python
plt.imshow(grayf, cmap='gray')
plt.title('Flipped Image')
```

```python
row,col = gray.shape
sft = gray.copy()
for r in range(row):
    for c in range(col-51):
        sft[r][c] = sft[r][c+40]
```

```python
plt.imshow(sft, cmap='gray')
plt.title('Shifted Image')
```

```python
rect = gray.copy()
cv2.rectangle(rect, (135,20),(225,110),(0,0,0),2)
plt.imshow(rect)
```

```python
import numpy as np
average= gray.copy()
average = average.astype(np.float64)
r , c = average.shape
avg_upd = average.copy()
for i in range(1,r-1):
    for j in range(1,c-1):
        avg_upd[i][j] = (average[i-1][j-1] + average[i-1][j] + average[i-1][j+1] + average[i][j-1] + average[i][j] + average[i][j+1] + average[i+1][j-1] + average[i+1][j] + average[i+1][j+1])/9
```

```python
plt.imshow(gray, cmap='gray')
plt.title('Original Image')
```

```python
plt.imshow(avg_upd, cmap='gray')
plt.title('Average Filter')
```

```python
simple_blur = cv2.blur(gray, (5,5))
gauss_blur = cv2.GaussianBlur(gray, (5,5), 0)
plt.imshow(simple_blur, cmap='gray')
plt.title('Simple Blur')
```

```python
plt.imshow(gauss_blur, cmap='gray')
plt.title('Gaussian Blur')
```
